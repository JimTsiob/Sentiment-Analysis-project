## Creation of Label column

### In this part we will create a Label column in order to test supervised learning methods

```python
# Function to map review ratings from overall column to labels
def label_rating(rating):
    if rating > 3:
        return 'POSITIVE'
    elif rating == 3:
        return 'NEUTRAL'
    elif rating < 3:
        return 'NEGATIVE'
    

electronics_dataset['Label'] = electronics_dataset['overall'].apply(label_rating)

electronics_dataset.head()
```
